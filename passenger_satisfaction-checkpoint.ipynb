# Some Useful Functions

Here I've defined a function for histogram for future uses. 

```python
def hist(feature, x , y):
    counts, bin_edges = np.histogram(feature, bins=10, density = True)
    pdf = counts/(sum(counts))
    print(pdf);
    print(bin_edges)

    #compute CDF
    cdf = np.cumsum(pdf)
    plt.plot(bin_edges[1:],pdf)
    plt.plot(bin_edges[1:], cdf)
    plt.gca().legend(('pdf','cdf'))
    plt.xlabel(x)
    plt.ylabel(y)

    plt.show();
```

Here, I'm going to do feature engineering based on the above results.
1. I am trying to build a feature with name <i> 'loy' </i> where in, if a person is Loyal and either travelling by Eco or Business or Eco Plus, The customer is considered as satisfied else the customer is dissatisfied.
2. I am trying to build another feature with name <i> 'tt' </i> where in, if a person is Loyal and either travelling by Business Travel or Personal Travel, The customer is considered as satisfied else the customer is dissatisfied.

```python

def loy(x,y):
    if (x=='Loyal Customer' and (y== 'Eco' or y== 'Business' or y== 'Eco Plus')) :
        return(1)
    else:
        return(0)
```

```python
df2['loy'] = list(map(loy,df2['Cus_Type'],df2['Class']))
```

```python
def tt(x,y):
    if (x=='Loyal Customer' and (y== 'Business Travel' or y== 'Personal Travel')) :
        return(1)
    else:
        return(0)
```

```python
df2['tt'] = list(map(tt,df2['Cus_Type'],df2['Travel_type']))
```

Now I'm importing necessary libraries and making necessary changes to my dataset to perform train-test split. <br>

df2 = train data + cross validate data + test data (i.e. total data set) <br>

# Random Forest Model

 Now I'm importing necessary libraries to build a model. I've also defined a function <i>"cm"</i> which will print my confusion matrix.

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score,f1_score,precision_score,accuracy_score
```

```python
def cm(df_cm):  
    """confusion_matrix_Plot = confusion_matrix(y_test,pred1)
    heatmap = sns.heatmap(confusion_matrix_Plot, annot=True,cmap="plasma", fmt="d")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right')
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right')
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
    tn, fp, fn, tp = df_cm.ravel()"""
    #df_cm = df_cm.astype('float') / df_cm.sum(axis=0)[1:, np.newaxis]
    sb.set(context='notebook',palette='deep')#for label size
    #for plot #https://matplotlib.org/examples/color/colormaps_reference.html
    sb.heatmap(df_cm, annot=True,cmap="Greens",xticklabels=["negative","positive"], yticklabels=["negative","positive"], fmt='g')
    plt.title('Confusion Matrix',fontsize=18)
    plt.xlabel('Predicted label',fontsize=15)
    plt.ylabel('True label',fontsize=15)
    plt.show()
    tn, fp, fn, tp = df_cm.ravel()
    print(" True Negative = "+str(tn)+"\n False positive = "+str(fp)+"\n False Negative = "+str(fn)+"\n True Positive = "+str(tp))
```

I'm building my model using Random Forest Classifier but with more number of estimators. Here alpha indicates number of estimators.

```python
alpha = [50,100,150,200,250,300,350,400]
max_depth = [5, 10]
accuracy_array = []
for i in alpha:
    for j in max_depth:
        print("for n_estimators =", i,"and max depth = ", j)
        clf = RandomForestClassifier(n_estimators=i, max_depth=j)
        clf.fit(xtrain,ytrain)
        #sig_clf = CalibratedClassifierCV(clf, method="sigmoid")
        #sig_clf.fit(X_train_,y_train_)
        pred = clf.predict(xcv) 
        #sig_clf_probs = sig_clf.predict_proba(X_cv_)
        accuracy_array.append(accuracy_score(ycv, pred))
        df_cm = confusion_matrix(ycv,pred)
        cm(df_cm)
 

print(accuracy_array)
best_alpha = np.argmax(accuracy_array)
clf = RandomForestClassifier(n_estimators=alpha[int(best_alpha/2)],max_depth=max_depth[int(best_alpha%2)])
clf.fit(xtrain,ytrain)


predict_y = clf.predict(xtrain)
print('For values of best estimator = ', alpha[int(best_alpha/2)], "The train accuracy score is:",accuracy_score(ytrain, predict_y))
predict_y = clf.predict(xcv)
print('For values of best estimator = ', alpha[int(best_alpha/2)], "The cross validation accuracy score is:",accuracy_score(ycv, predict_y))
```

<h3>4.5.2. Testing model with best hyper parameters (One Hot Encoding)</h3>

```python
clf = RandomForestClassifier(n_estimators=alpha[int(best_alpha/2)], criterion='gini', max_depth=max_depth[int(best_alpha%2)], random_state=42, n_jobs=-1)
clf.fit(xtr,ytr)
pred = clf.predict(xtest) 
accuracy_array.append(accuracy_score(ytest, pred))
df_cm = confusion_matrix(ytest,pred)
cm(df_cm)
predict_y = clf.predict(xtest)
print('For values of best estimator = ', alpha[int(best_alpha/2)], "The test accuracy score is:",accuracy_score(ytest, predict_y))
```
